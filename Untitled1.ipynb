# split dataset & querys

```python
from nltk.stem.porter import PorterStemmer
myfile = open("cran.all.1400",encoding = "UTF-8")
text = myfile.read()
m=text.split('.I')
stemmer =PorterStemmer()
tokenized_documents = [[token for token in document.lower().split()] for document in m]

dictionary = corpora.Dictionary(tokenized_documents)
v=[]
for i in range(len(tokenized_documents)):
    tokens_clean=[t for t in tokenized_documents[i] if t not in stop_words]
    #print('__________\n',tokens_clean)
    # remove all tokens that are not alphabetic
    words = [word for word in tokens_clean if word.isalpha()]
    stems=[stemmer.stem(t) for t in words ]
    v.append(stems)
dictionary = corpora.Dictionary(v)
print(v)
#print(dictionary.token2id)  
corpus_doc2bow_vectors = [dictionary.doc2bow(tok_doc) for tok_doc in v] 
#print(corpus_doc2bow_vectors)
tfidf_model= models.TfidfModel(corpus_doc2bow_vectors,id2word=dictionary,normalize=False)
corpus_doc2bow_vectors=tfidf_model[corpus_doc2bow_vectors]    
```

```python
index_metrix= similarities.SparseMatrixSimilarity(corpus_doc2bow_vectors,num_features=100000)
```

```python
def query(query):
        #qurey 
        tokens_clean1=query.lower().split()
        tokens_clean1=[t for t in tokens_clean1 if t not in stop_words]
        #print('__________\n',tokens_clean1)
        # remove all tokens that are not alphabetic
        words1 = [word for word in tokens_clean1 if word.isalpha()]
        stems1=[stemmer.stem(t) for t in words1 ]
        print(stems1)
        #similarity
        query_doc2bow_vectors = dictionary.doc2bow(stems1)
        #print(corpus_doc2bow_vectors)
        quert_tfidf_vector=tfidf_model[query_doc2bow_vectors]   
        sims =index_metrix[quert_tfidf_vector]
        l=sorted(list(enumerate(sims)),key = lambda tup:tup[1],reverse=True)
        return l

    
   
    
    
  
```
